<h1>Merging order with product Weight


```python
order_with_weight =pd.merge(df1,df2,on='SKU')
order_with_weight
```

<h1> 2.Calculate total weight per order(in grams)

```python
order_weight = order_with_weight.groupby('ExternOrderNo')["Weight (g)"].sum().reset_index()
order_weight['total weight (KG)'] =order_weight['Weight (g)']/1000
```

```python
order_weight
```

<h1>3.CALCULATE WEIGHT SLABS

```python
import math
def weight_slab(weight):
    if weight <= 0.5:
        return 0.5
    return math.ceil(weight * 2) / 2 

order_weight['Weight Slab'] = order_weight['total weight (KG)'].apply(weight_slab)
```

```python
order_weight
```

<h1>6 CALCULATE EXPECT CHARGES

```python
final_data = final_data.rename(columns={
    'Zone_x': 'Your_Calculated_Zone', 
    'Zone_y': 'Courier_Claimed_Zone'  
})

final_data
```

```python
def calculate_expected_charge(row):
    zone = row['Your_Calculated_Zone'].lower()  
    weight_slab = row['Weight Slab']
    shipment_type = row['Type of Shipment']
    
    try:
        if shipment_type == 'Forward charges':
            fixed = df5[f'fwd_{zone}_fixed'].iloc[0]
            additional = df5[f'fwd_{zone}_additional'].iloc[0]
        else: 
            fixed = df5[f'fwd_{zone}_fixed'].iloc[0] + df5[f'rto_{zone}_fixed'].iloc[0]
            additional = df5[f'fwd_{zone}_additional'].iloc[0] + df5[f'rto_{zone}_additional'].iloc[0]
        
        extra_slabs = max(0, (weight_slab - 0.5) / 0.5)
        return fixed + (additional * extra_slabs)
    
    except KeyError:
        print(f"Rates not found for zone: {zone}")
        return 0  
final_data['Expected_Charge'] = final_data.apply(calculate_expected_charge, axis=1)
```

```python
final_data
```

<h1>7 COMPARE CHARGEES

```python
final_data['Charge_Difference'] = final_data['Expected_Charge'] - final_data['Billing Amount (Rs.)']

# # Categorize orders
final_data['Billing_Status'] = 'Corr234ect'
final_data.loc[final_data['Charge_Difference'] < 0, 'Billing_Status'] = 'Overcharged'
final_data.loc[final_data['Charge_Difference'] > 0, 'Billing_Status'] = 'Undercharged'
```

```python
final_data
```

<h1>CREATE SUMMARY

```python
# Group by Billing_Status and aggregate
ssummary = final_data.groupby('Billing_Status').agg(
    Order_Count=('Order ID', 'count'),
    Amount=('Charge_Difference', 'sum')
).reset_index()

# Adjust the Amount for 'Overcharged' and 'Undercharged' statuses
ssummary.loc[ssummary['Billing_Status'] == 'Overcharged', 'Amount'] = abs(ssummary.loc[ssummary['Billing_Status'] == 'Overcharged', 'Amount'])
ssummary.loc[ssummary['Billing_Status'] == 'Undercharged', 'Amount'] = abs(ssummary.loc[ssummary['Billing_Status'] == 'Undercharged', 'Amount'])

# Rename Billing_Status to Category
ssummary['Category'] = ssummary['Billing_Status'].replace({
    'Correct': 'Total orders where X has been correctly charged',
    'Overcharged': 'Total Orders where X has been overcharged',
    'Undercharged': 'Total Orders where X has been undercharged'
})

# Reorder columns
ssummary = ssummary[['Category', 'Order_Count', 'Amount']]

# Display the summary
print(ssummary)
```
